**สร้างกราฟ Elblow method และ KMEAN clastering**


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Method 1: Double Backslashes
file_path = 'C:\\Users\\ACER\\Desktop\\train.csv'  # Update the path to your CSV file

# Method 2: Raw String
# file_path = r'C:\Users\ACER\Desktop\train.csv'  # Update the path to your CSV file

# Method 3: Forward Slashes
# file_path = 'C:/Users/ACER/Desktop/train.csv'  # Update the path to your CSV file

# Load the dataset
data = pd.read_csv('train.csv')

# Selecting the necessary columns
data_selected = data[['total product', 'Purchase']]

# Handling missing values by dropping rows with any NaN values
data_selected = data_selected.dropna()

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

# Determining the optimal number of clusters using the Elbow Method
inertia = []
range_n_clusters = range(1, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, inertia, marker='o')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Training KMeans with the chosen number of clusters (e.g., k=3)
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(data_scaled)

# Adding the cluster information to the original data
data_selected['Cluster'] = clusters

# Plotting the clusters
plt.figure(figsize=(10, 8))
plt.scatter(data_selected['total product'], data_selected['Purchase'], c=clusters, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('Total Product')
plt.ylabel('Purchase')
plt.colorbar(label='Cluster')
plt.show()
```

## Cluster แปลเป็นกลุ่มลูกค้าที่แตกต่างกัน
Cluster 0: ลูกค้าที่มี total product และ Purchase ต่ำ

Cluster 1: ลูกค้าที่มี total product และ Purchase สูง

Cluster 2: ลูกค้าที่มีพฤติกรรมการซื้อและ total product ปานกลาง

##Insight for Business

ลูกค้าที่มีมูลค่าสูง (Cluster 1): มุ่งเน้นการรักษาลูกค้าเหล่านี้ด้วย Promotion และservice
ลูกค้าที่มีมูลค่าต่ำ (Cluster 0): เน้นเพื่อเพิ่มการมีส่วนร่วมและปริมาณการซื้อของพวกเขา เช่น discount หรือผลิตภัณฑ์ระดับเริ่มต้น
ลูกค้าที่มีมูลค่าปานกลาง (Cluster 2): ระบุปัจจัยที่อาจเปลี่ยนพวกเขาให้เป็นลูกค้าที่มีมูลค่าสูง
